```python
# Search by Plant Name
# Search functions
def search_by_name(plant_name):
    result = data[data['Common Name(s)'].str.contains(plant_name, case=False, na=False) | 
                  data['Botanical Name'].str.contains(plant_name, case=False, na=False)]
    return result if not result.empty else "No results found."
```

```python
# Search by Application Keyword
def search_by_application(keyword):
    result = data[data['Medicinal Uses'].str.contains(keyword, case=False, na=False) | 
                  data['Active Compounds'].str.contains(keyword, case=False, na=False) | 
                  data['Method of Preparation'].str.contains(keyword, case=False, na=False)]
    return result if not result.empty else "No results found."
```

```python
from flask import Flask, render_template, request
import pandas as pd

app = Flask(__name__)

# Load the dataset (Excel file)
data = pd.read_excel('Medical_plants1.xlsx')

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/search', methods=['POST'])
def search():
    search_term = request.form['search_term']
    search_option = request.form['search_option']
    
    # Logic to filter based on search option
    if search_option in ['Botanical Name', 'Common Name(s)']:
        result = data[data[search_option].str.contains(search_term, case=False, na=False)]
    elif search_option in ['Medicinal Uses', 'Active Compounds', 'Method of Preparation']:
        result = data[data[search_option].str.contains(search_term, case=False, na=False)]
        result = result[['Plant_Name', 'Medicinal_Uses', 'Active_Compounds', 'Method_of_Preparation']]
    else:  # For Parts Used and Geographical Distribution
        result = data[data[search_option].str.contains(search_term, case=False, na=False)]
        result = result[['Plant_Name', search_option]]

    return render_template('result.html', results=result)

if __name__ == '__main__':
    app.run(port=5000)


app.run(port=5000, use_reloader=False)
```
